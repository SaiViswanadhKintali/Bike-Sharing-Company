# Multiple Linear Regression

## Bike Sharing Assignment

### Problem Statement
The Bike Sharing compant want to understand the factors on which the demand for these shared bikes depends. Specifically, they want to understand the factors affecting the demand for these shared bikes in the American market. The company wants to know:

- Which variables are significant in predicting the demand for shared bikes.
- How well those variables describe the bike demands

### Business Goal
Model the demand for shared bikes with the available independent variables. It will be used by the management to understand how exactly the demands vary with different features. This should help them manipulate the business strategy to meet the demand levels and meet the customer"s expectations.

#### Step 3: Data Preparation

```python
# Creating Dummies

# cat_cols_for_dummies = ["season","mnth","weekday","weathersit","day",]
# cat_cols_binary = ["yr","holiday","workingday"]

df_dum = pd.get_dummies(df,drop_first=True)
df_dum.head()
```

```python
df_dum.describe()
```

```python
df_dum[["temp","atemp","hum","windspeed","casual","registered","cnt"]] = df_dum[["temp","atemp","hum","windspeed","casual","registered","cnt"]].astype(float)
```

```python
df_dum.info()
```

```python
from sklearn.model_selection import train_test_split

np.random.seed(0)
df_train, df_test = train_test_split(df_dum, train_size = 0.7, test_size = 0.3, random_state = 64)
```

Rescaling the Features

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
cols = ["temp","atemp","hum","windspeed","casual","registered","cnt"]
df_train[cols] = scaler.fit_transform(df_train[cols])
df_train
```

```python
# Diving X & y for model building.

y_train = df_train.pop("cnt")
X_train = df_train
```

#### Step 5: Building a linear model

```python
import statsmodels.api as sm

# Add a constant
X_train_lm = sm.add_constant(X_train)

# Create a first fitted model
lr = sm.OLS(y_train, X_train_lm).fit()
lr.params
```

```python
print(lr.summary())
```

```python
# Checking VIF
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
# Checking VIF
vif = pd.DataFrame()
vif['Features'] = X_train.columns
vif['VIF'] = [variance_inflation_factor(X_train.values, i) for i in range(X_train.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```
